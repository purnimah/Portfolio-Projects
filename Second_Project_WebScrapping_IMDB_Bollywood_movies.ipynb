```python
movie_list = soup.find_all('div', class_="lister-item mode-advanced") #selecting the main div so that we can extract data in a loop.
```

```python
len(movie_list) 
```

```python
year = []
rating = []
number_of_votes = []
genre = []
runtime = []
director_cast = []

for i in movie_list:
    year.append(i.find('span', class_="lister-item-year text-muted unbold").text.strip())
    rating.append(i.find('strong').text.strip())
    number_of_votes.append(i.find('span', {"name":"nv"}).text.strip())
    genre.append(i.find('span', class_="genre").text.strip())
    runtime.append(i.find('span', class_="runtime").text.strip())
    director_cast.append(i.find('p',class_="").text.replace('\n',""))
   
```

```python
d={'Movie Name':movies, 'Year of Release':year, 'Rating':rating, 'Genre':genre, 'Runtime':runtime, 'No. of Voters':number_of_votes, 'Director and Cast':director_cast}
```

### _This is the data extracted from the first page now we have to do the same for the rest of the pages_

```python
movie_list[0].find_all('a')[1].text 
```

This is another way of geeting just the title as the 2nd element (1st consideing the list as indexing starts from 0) will always be the title. The same menthod will not work for the name of the director and stars are there is no gurantee that the director element will be on the same positon as there can be multiple direcotrs hence multiple anchor tag and we also don't have a specific class of anchor tag to extract from, so indexing is the only option. To reduce the code I'll be using the indexing technique to create the final data frame of 500 movie list

```python
df_new = pd.DataFrame()

for j in range (1,501,50):
    url = 'https://www.imdb.com/search/title/?title_type=feature&primary_language=hi&sort=num_votes,desc&start=51&ref_=adv_prv'
    req = 'https://www.imdb.com/search/title/?title_type=feature&primary_language=hi&sort=num_votes,desc&start=' + str(j) + '&ref_=adv_prv'
    webpage = requests.get(req).text
    soup = BeautifulSoup(webpage, 'lxml')
    movie_list = soup.find_all('div', class_="lister-item mode-advanced") #selecting the main div so that we can extract data in a loop.
    name = []
    year = []
    rating = []
    number_of_votes = []
    genre = []
    runtime = []
    director_cast = []

    for i in movie_list:
        name.append(i.find_all('a')[1].text)
        year.append(i.find('span', class_="lister-item-year text-muted unbold").text.strip())
        rating.append(i.find('strong').text.strip())
        number_of_votes.append(i.find('span', {"name":"nv"}).text.strip())
        genre.append(i.find('span', class_="genre").text.strip())
        runtime.append(i.find('span', class_="runtime").text.strip())
        director_cast.append(i.find('p',class_="").text.replace('\n',""))
   
    d2={'Movie Name': name, 'Year of Release':year, 'Rating':rating, 'Genre':genre,'Runtime':runtime, 'No. of Voters':number_of_votes, 'Director and Cast':director_cast}
    df2=pd.DataFrame(d2)
    df_new=df_new.append(df2)
```
